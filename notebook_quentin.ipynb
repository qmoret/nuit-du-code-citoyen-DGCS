```python
# Nombre moyen de nuits par utilisateur
demandes = df.groupby(['per_id','dem_id'],as_index=0)['rep_nb_nuit'].mean()
demandes = demandes[~demandes['rep_nb_nuit'].isnull()]
demandes.groupby(['per_id'])['rep_nb_nuit'].sum().mean()
```

```python
# Quel est le temps moyen entre deux demandes par utilisateur ? # il faudrait l'avoir par demandeur principal
demandes = df[['per_id','dem_date']].drop_duplicates()
dem_min = demandes.groupby(['per_id'],as_index=0)['dem_date'].min()
dem_max = demandes.groupby(['per_id'],as_index=0)['dem_date'].max()
dem_count = demandes.groupby(['per_id'],as_index=0).agg({'dem_date':'nunique'})

demandes_dates = dem_min.join(dem_max,rsuffix='_max').join(dem_count,rsuffix='_count')
demandes_dates['date_diff'] = demandes_dates['dem_date_max'] - demandes_dates['dem_date']
demandes_dates['ecart_moy'] = demandes_dates.date_diff / demandes_dates['dem_date_count']
demandes_dates.ecart_moy.mean()
```

```python
# Pourcentage de d'utilisateurs faisant seulement une demande
(demandes_dates.date_diff==timedelta(0)).mean()
# 37.5%
```
